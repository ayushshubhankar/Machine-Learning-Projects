# Machine Learning

* We would split the data into train and test.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.25,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

* After splitting the data into training and test, would then use StandardScaler on the training data.

* Using standardscaler so as to arrange the data in form of standard normal distribution.

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train_std=sc.fit_transform(x_train)
x_test_std= sc.transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

* Now we would be following the machine learning pathway and would be implementing linear regression.
The pathcould be accessed at https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html

```python
from sklearn import linear_model
lr= linear_model.LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred= lr.predict(x_test)
```

```python
y_pred
```

```python
y_pred_reshape= np.reshape(y_pred, (len(y_pred),1))
y_pred_reshape
```

* Now we would be printing accuracy scores using coefficient of determination, Mean Sqaured Error and mean absolute error.

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
r_square=r2_score(y_test,y_pred)
```

```python
r_square
```

```python
mae= mean_absolute_error(y_test,y_pred)
mae
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
accuracy= {"Method":["Coefficient of Determination","Mean Sqaured Error" ,"Mean Absolute Error"],"Score":[float(r_square), float(mse),float(mae)] }
pd.DataFrame.from_dict(accuracy)
```
