# Machine Learning

```python
#Splitting the Data into X and Y. 
# X would contain all the cloumns except Name and Status.
x=df.drop(columns=['name','status'],axis=1)
# Y would only contain the Loan Status.
y=df["status"]
```

* We would be following the steps as per the pathway at https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html
* Since this is a classification problem, we would be following the classification pathway

```python
# Splitting the DATA into Training and Test with a test size of 20%
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2)
```

```python
y_test
```

```python
# Scaling the data by using Standard Scaler
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train= sc.fit_transform(x_train)
x_test= sc.transform(x_test)
```

### Linear SVC

```python
# Importing the Linear SVC from the Scikit Learn Library.
from sklearn.svm import LinearSVC
lc=LinearSVC()
lc.fit(x_train,y_train)
y_pred_lc= lc.predict(x_test)
y_test
```

```python
from sklearn.metrics import accuracy_score
lc_score= round((accuracy_score(y_pred_lc,y_test))*100)
print("The Accuracy score for Linear SVC is ", lc_score, "%")
```

### KNN Method

```python
# Importing thr Libraries
from sklearn.neighbors import KNeighborsClassifier
# Gathering the optiumum number of Neighbors
for i in range(1,15):
    kn=KNeighborsClassifier(n_neighbors=i)
    kn.fit(x_train,y_train)
    y_pred_kn= kn.predict(x_test)
    kn_score=round((accuracy_score(y_pred_kn,y_test)*100))
    print(i, "-----", kn_score,"%")
    i=i+1


```

```python
# Predicting the scores using the Optimim Neighbors.
kn=KNeighborsClassifier(n_neighbors=6)
kn.fit(x_train,y_train)
y_pred_kn= kn.predict(x_test)
kn_score=round((accuracy_score(y_pred_kn,y_test)*100))
print("The Accuracy score for KNN is ", kn_score, "%")
```

### Random Forest

```python
# Importing the Libraries.
from sklearn.ensemble import RandomForestClassifier
# Gathering the optiumum number of estimators
for i in range(1,150):
    rf=RandomForestClassifier(n_estimators=i)
    rf.fit(x_train,y_train)
    y_pred_rf= rf.predict(x_test)
    rf_score=round((accuracy_score(y_pred_rf,y_test)*100))
    print(i, "-----", rf_score,"%")
    i=i+1
```

```python
# Predicting the scores using the Optimim Estimators.
rf= RandomForestClassifier(n_estimators=63)
rf.fit(x_train,y_train)
y_pred_rf= rf.predict(x_test)
rf_score=round((accuracy_score(y_pred_rf,y_test)*100))
print("The Accuracy score for RandomForest is ", rf_score, "%")
```

```python
# Concatenating the Scores of Various Models.
scores= {"Model":["Linear SVC","KNN Method","RandomForrest"], "Scores":[lc_score,kn_score,rf_score]}
scores= pd.DataFrame.from_dict(scores)
scores
```

```python
# Plotting the Data for Visualisation
sns.barplot(x="Model",y="Scores",data= scores)
plt.title("Scores Comparison",weight="bold",size=15)
```
