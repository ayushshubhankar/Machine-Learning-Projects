# Visual Comparison

```python
#setting the Layout for side by side Viewing
fig,(ax1,ax2,ax3)=plt.subplots(nrows=1, ncols=3, figsize=(15,5))
fig.suptitle("Content Comparison", weight="bold",color="black",size=30)
fig.tight_layout(pad=2)
sns.set_theme(style="whitegrid")

# plotting the content types of Amazon Prime
fig1 = sns.countplot(x="type",data=am,ax=ax1,hue="type",palette="tab10")
fig1.set_xlabel("Type")
fig1.set_ylabel("Count")
fig1.set_title("Amazon Prime",weight="bold",size=15)

# plotting the content types of Hotstar
fig2 = sns.countplot(x="type",data=hot,ax=ax2,hue="type",palette="tab10")
fig2.set_xlabel("Type")
fig2.set_ylabel("Count")
fig2.set_title("Hotstar",weight="bold",size=15)

# plotting the content types of Netflix
fig3 = sns.countplot(x="type",data=net,ax=ax3,hue="type",palette="tab10")
fig3.set_xlabel("Type",weight="bold",size=15)
fig3.set_ylabel("Count")
fig3.set_title("Netflix",weight="bold",size=15)
```

### Inference
* All three Platforms have more Number of Movies in comaprison to TV Shows.
* Amazon Prime has fairly less TV shows in comparison to Movies.
* Number of Movies in Hotstar is more than that of Amazon prime.
* Netflix has the Most number of Movies amongst all of them.

```python
#setting the Layout for side by side Viewing
fig,(ax1,ax2,ax3)=plt.subplots(nrows=1, ncols=3, figsize=(15,5))
fig.suptitle("Yearwise Release", weight="bold",color="black",size=30)
fig.tight_layout(pad=2)
sns.set_theme(style="darkgrid")

# plotting the yearwise release of Amazon Prime
fig1 = sns.stripplot(x="release_year", y="type",data=am,ax=ax1,palette="tab10")
fig1.set_xlabel("Year of Release")
fig1.set_ylabel("Type")
fig1.set_title("Amazon Prime",weight="bold",size=15)

# plotting the yearwise release of Hotstar
fig2 = sns.stripplot(x="release_year", y="type",data=hot,ax=ax2,palette="tab10")
fig2.set_xlabel("Year of Release")
fig2.set_ylabel("Type")
fig2.set_title("Hotstar",weight="bold",size=15)

# plotting the yearwise release of Netflix
fig3 = sns.stripplot(x="release_year", y="type",data=net,ax=ax3,palette="tab10")
fig3.set_xlabel("Year of Release")
fig3.set_ylabel("Type")
fig3.set_title("Netflix",weight="bold",size=15)
```

### Inference
* All three Platforms host recent TV shows in comaprison to Movies.
* Amazon Prime hosts one of the oldest movies and TV Shows on its platform.
* Movies on Hotstar are of similar age in comaprison to Amazon Prime. 
* Netflix hosts large number of Movies which have been released recently.

```python
#plotting the category wise release on Amazon Prime
fig, ax = plt.subplots(figsize=(25,10))
am_pal=sns.color_palette("Paired")
am_fig=sns.countplot(x="rating", data=am,palette=am_pal,dodge=False,hue="type",)
am_fig.legend(loc='upper right')
am_fig.set_xlabel("Year of Release",weight="bold",size=20)
am_fig.set_ylabel("Type",weight="bold",size=20)
am_fig.set_title("Category Wise Release on Amazon Prime",weight="bold",size=30)


```

```python
#plotting the category wise release on Hotstar
fig, ax = plt.subplots(figsize=(20,10))
sns.set_theme(style="whitegrid")
hot_pal=sns.color_palette("Paired")
am_fig=sns.countplot(x="rating", data=hot,palette=hot_pal,dodge=False,hue="type")
am_fig.legend(loc='upper right')
am_fig.set_xlabel("Year of Release",weight="bold",size=20)
am_fig.set_ylabel("Type",weight="bold",size=20)
am_fig.set_title("Category Wise Release on Hotstar",weight="bold",size=30)
```

```python
# plotting the category wise release on Netflix
fig, ax = plt.subplots(figsize=(25,10))
net_pal=sns.color_palette("Paired")
sns.set_theme(style="whitegrid")
am_fig=sns.countplot(x="rating", data=am,palette=net_pal,dodge=False,hue="type")
am_fig.legend(loc='upper right')
am_fig.set_xlabel("Year of Release",weight="bold",size=20)
am_fig.set_ylabel("Type",weight="bold",size=20)
am_fig.set_title(" Category Wise Release on Netflix",weight="bold",size=30)
```

### Inference
* All three Platforms have more Number of Movies in comaprison to TV Shows, in each Category.
* Amazon Prime has similar category wise releases in comaprison to Netflix.

* Hotstar Offers the least Number of Categories amongst all of them.
